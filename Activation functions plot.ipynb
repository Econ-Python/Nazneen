```python
import numpy as np
import matplotlib.pyplot as plt

# Define the ReLU and Softmax activation functions
def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Generate input values for ReLU
x = np.linspace(-10, 10, 400)

# Apply the ReLU and tanh activation function
relu_values = relu(x)
tanh_values = tanh(x)

# Plot the functions
plt.figure(figsize=(10, 6))

# ReLU plot
plt.subplot(1, 2, 1)
plt.plot(x, relu_values, label="ReLU", color='b')
plt.title("ReLU Activation Function")
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.legend()

# Tanh plot
plt.subplot(1, 2, 2)
plt.plot(x, tanh_values, label="Tanh", color='r')
plt.title("Tanh Activation Function")
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
relu = lambda x: np.maximum(0, x)
tanh = np.tanh

# Generate input values
x = np.linspace(-10, 10, 400)

# Create the plot
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Plot ReLU and Tanh
for i, (func, color, label) in enumerate([(relu, 'b', 'ReLU'), (tanh, 'r', 'Tanh')]):
    ax[i].plot(x, func(x), color=color, label=label)
    ax[i].set_title(f"{label} Activation Function")
    ax[i].set_xlabel("x")
    ax[i].set_ylabel(f"{label}(x)")
    ax[i].grid(True)
    ax[i].legend()
    ax[i].spines['left'].set_position(('data', 0))  # Move y-axis to x=0
    ax[i].spines['bottom'].set_position(('data', 0))  # Move x-axis to y=0
    ax[i].xaxis.set_ticks_position('bottom')  # Ensure ticks are at the bottom
    ax[i].yaxis.set_ticks_position('left')  # Ensure ticks are at the left

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the activation functions
relu = lambda x: np.maximum(0, x)
sigmoid = lambda x: 1 / (1 + np.exp(-x))
tanh = np.tanh
leaky_relu = lambda x: np.maximum(0.01 * x, x)
softmax = lambda x: np.exp(x) / np.sum(np.exp(x), axis=0)  # Softmax for a vector
elu = lambda x, alpha=1: np.where(x > 0, x, alpha * (np.exp(x) - 1))
swish = lambda x: x * sigmoid(x)

# Generate input values
x = np.linspace(-10, 10, 400)

# Generate logits for Softmax (e.g., 5 classes)
logits = np.linspace(-3, 3, 5)

# Create the plot
fig, ax = plt.subplots(2, 4, figsize=(18, 10))

# Define the functions and their labels
functions = [
    (relu, 'ReLU', 'b'),
    (sigmoid, 'Sigmoid', 'g'),
    (tanh, 'Tanh', 'r'),
    (leaky_relu, 'Leaky ReLU', 'c'),
    (softmax, 'Softmax', 'm'),
    (elu, 'ELU', 'y'),
    (swish, 'Swish', 'purple')
]

# Plot the functions
for i, (func, label, color) in enumerate(functions):
    row, col = divmod(i, 4)  # Plotting in a 2x4 grid
    if label == 'Softmax':
        softmax_values = softmax(logits)
        ax[row, col].bar(range(len(logits)), softmax_values, color=color, tick_label=[f"Class {i+1}" for i in range(len(logits))])
        ax[row, col].set_ylim(0, 1)
        ax[row, col].set_ylabel(f"{label} Probability")
    else:
        ax[row, col].plot(x, func(x), color=color, label=label)
        ax[row, col].set_ylabel(f"{label}(x)")

    ax[row, col].set_title(f"{label} Activation Function")
    ax[row, col].set_xlabel("x" if label != 'Softmax' else "Class")
    ax[row, col].grid(True)
    ax[row, col].legend()
    ax[row, col].spines['left'].set_position(('data', 0))  # Move y-axis to x=0
    ax[row, col].spines['bottom'].set_position(('data', 0))  # Move x-axis to y=0
    ax[row, col].xaxis.set_ticks_position('bottom')  # Ensure ticks are at the bottom
    ax[row, col].yaxis.set_ticks_position('left')  # Ensure ticks are at the left

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as pt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / exp_x.sum(axis=0)

x = np.linspace(-10, 10, 100)
y_sig = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)
y_sm = softmax(x)

pt.figure(figsize=(12, 8))

pt.subplot(2, 2, 1)
pt.plot(x, y_sig, color='blue', label='Sigmoid')
pt.title("Sigmoid")
pt.xlabel("x")
plt.ylabel("sigmoid(x)")
pt.grid(True)
pt.legend()

pt.subplot(2, 2, 2)
pt.plot(x, y_tanh, color='green', label='Tanh')
pt.title("tanh")
pt.xlabel("x")
pt.ylabel("tanh(x)")
pt.grid(True)
pt.legend()

pt.subplot(2, 2, 3)
pt.plot(x, y_relu, color='red', label='ReLU')
pt.title("ReLU")
pt.xlabel("x")
pt.ylabel("ReLU(x)")
pt.grid(True)
pt.legend()

x_sm = np.linspace(-2, 2, 100)
y_sm = softmax(x_sm)
pt.subplot(2, 2, 4)
pt.plot(x_sm, y_sm, color='purple', label='Softmax')
pt.title("Softmax")
pt.xlabel("x")
pt.ylabel("softmax(x)")
pt.grid(True)
pt.legend()

pt.tight_layout()
pt.show()
```
